###1.2.5 Тепловая карта корреляций

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Первая тепловая карта (корреляция)
corr_matrix = df.corr()

# Вторая тепловая карта (агрегированные значения)
df['quality_cat'] = pd.cut(
    df['quality'],
    bins=[0, 5, 7, 10],
    labels=['low', 'medium', 'high']
)

pivot_table = df.pivot_table(
    values='alcohol',
    index='quality',
    columns='quality_cat',
    aggfunc='mean'
)

# Объединённый вывод
plt.figure(figsize=(16, 6))

# Первая heatmap
plt.subplot(1, 2, 1)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляция признаков')

# Вторая heatmap
plt.subplot(1, 2, 2)
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title('Средний alcohol по категориям качества')

plt.tight_layout()
plt.show()
```

###1.2.6 Удаление дубликатов строк

###1.2.7 Появление нетипичных выбросов по признакам

###1.2.8. Условная фильтрация сэмплов

```python
# 1. Фильтрация по одному признаку: высокий alcohol
df_high_alcohol = df[df['alcohol'] > 12]

print("Вина с alcohol > 12:", df_high_alcohol.shape[0])
df_high_alcohol.head()

# 2. Фильтрация по нескольким признакам: высокое качество и alcohol выше 11
df_high_quality_alcohol = df[
    (df['quality'] >= 7) & (df['alcohol'] > 11)
]

print("Вина с quality >= 7 и alcohol > 11:", df_high_quality_alcohol.shape[0])
df_high_quality_alcohol.head()

# 3. Фильтрация по диапазону: кислотность в умеренном диапазоне
df_moderate_acidity = df[
    (df['pH'] >= 3.0) & (df['pH'] <= 3.4)
]

print("Вина с pH от 3.0 до 3.4:", df_moderate_acidity.shape[0])
df_moderate_acidity.head()

# 4. Для наглядного сравнения размеров полученных выборок
filter_counts = {
    'alcohol > 12': df_high_alcohol.shape[0],
    'quality >= 7 и alcohol > 11': df_high_quality_alcohol.shape[0],
    '3.0 <= pH <= 3.4': df_moderate_acidity.shape[0]
}

filter_counts
```

```python
plt.figure(figsize=(12, 4))

# Первый график
plt.subplot(1, 2, 1)
sns.histplot(x='quality', y='alcohol', data=df_high_alcohol)
plt.title('Качество vs Alcohol (alcohol > 12)')

#разделитель
print ('')

# Второй график
plt.subplot(1, 2, 2)
sns.histplot(x='density', y='sulphates', data=df_high_alcohol)
plt.title('Density vs Sulphates (alcohol > 12)')

plt.tight_layout()
plt.show()
```

###1.2.9. Добавление шума

```python
import numpy as np

np.random.seed(42)

plt.title('Сравнение распределений alcohol до и после добавления шума')
df['alcohol_noise'] = df['alcohol'] + np.random.normal(0, 0.1, len(df))
sns.kdeplot(df['alcohol'], label='original')
sns.kdeplot(df['alcohol_noise'], label='noise')

plt.legend()
plt.show()
```
